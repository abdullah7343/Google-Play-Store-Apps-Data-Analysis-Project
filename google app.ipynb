```python
null_var=df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_col=null_var[null_var>10].keys()
drop_col
```

```python
drop=df.drop(columns=drop_col,axis=1,inplace=True)
df.shape
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.isnull())
```

```python
df['Type'].fillna(df['Type'].mode()[0],inplace=True)
df['Android Ver'].fillna(df['Android Ver'].mode()[0],inplace=True)
df['Current Ver'].fillna(df['Current Ver'].mode()[0],inplace=True)
df['Content Rating'].fillna(df['Content Rating'].mode()[0],inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.isnull())
```

```python
df.info()
```

```python
df.head()
```

```python
df["Reviews"] = pd.to_numeric(df["Reviews"], errors="coerce")
```

```python
df["Installs"] = df["Installs"].str.replace("+","")
df["Installs"] = df["Installs"].str.replace(",","")
df["Installs"] = pd.to_numeric(df["Installs"], errors="coerce")
```

```python
df["Price"] = df["Price"].str.replace("$","")
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"], errors="coerce")
```

```python
df.dtypes
```

```python
df['Category'].value_counts()
```

```python
df = pd.get_dummies(df, columns=["Category"], dtype=int)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df["Type"] = le.fit_transform(df["Type"].astype(str))
df["Content Rating"] = le.fit_transform(df["Content Rating"].astype(str))
df["Genres"] = le.fit_transform(df["Genres"].astype(str))
df["Android Ver"] = le.fit_transform(df["Android Ver"].astype(str))
```
